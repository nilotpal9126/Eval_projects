```python
from scipy.stats import zscore
import numpy as np
```

```python
z=np.abs(zscore(df))
print(z)
treshhold=3 
print (np.where(z<3))
df_new=df[(z<3).all(axis=1)]
print(df_new)
df_new.shape
```

### Splitting the data

```python
from sklearn.model_selection import train_test_split
```

```python
y= df_new['W']
print(y, y.shape)
x=df_new.drop(['W'], axis=1)
print(x,x.shape)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.22, random_state=22)
print(x_train,x_test,y_train,y_test)
```

### Standardising the trainning values

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
print(scaler.fit(x_train,y_train))
```

##### Model training / Hype-parameter tuning/ Checking model accuracy

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
print(lr.score(x_train,y_train))
lr_pred=lr.predict(x_test)
y_pred=lr.predict(x_test)
print(lr_pred)
from sklearn.linear_model import Lasso,Ridge
ls=Lasso(alpha=0.0001)
ls.fit(x_train,y_train)
print(ls.score(x_train,y_train))
print(lr.score(x_test, y_test))
```

```python
ls.coef_
```

```python
rd=Ridge(0.001)
rd.fit(x_train,y_train)
rd.score(x_train,y_train)
```

```python
rd.coef_
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(np.sqrt(mean_squared_error(y_test,y_pred)))
```

These error values show a less score so it can be safe to assume that the model holds good with a lr accuacy score of 68%.

```python
plt.scatter(y_test,lr_pred)
plt.xlabel('actual average_price')
plt.ylabel('predicted average_price ')
plt.title('actual vs predicted model')
plt.show()
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,lr_pred))

#A high r2 score shows that 68.98% of the data data fit the model.
```
