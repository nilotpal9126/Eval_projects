##### splitiing into x and y

```python
x=train_df.drop(['Fees'], axis=1)
y=train_df['Fees']
```

```python
from sklearn.preprocessing import StandardScaler
X=StandardScaler().fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression  
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingRegressor,RandomForestRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
from sklearn.linear_model import Lasso,Ridge
ls=Lasso(alpha=0.0001)
ls.fit(X_train,y_train)
ls.score(X_train,y_train)
print('ls_coef',ls.coef_)
rd=Ridge(0.001)
rd.fit(X_train,y_train)
print('rd_score',rd.score(X_train,y_train))
print('rd_coef',rd.coef_)
```

```python
models=[LinearRegression(),KNeighborsRegressor(n_neighbors=5),DecisionTreeRegressor(),RandomForestRegressor(n_estimators=100,max_features='auto'),GradientBoostingRegressor()]
for m in models:
    m.fit(X_train,y_train)
    y_pred=m.predict(X_test)
    score=m.score(X_train,y_train)
    pred=m.predict(X_test)
    print(score,m,':')
    print("Predictions are: {}".format(pred))
    
    r2score=r2_score(y_test,pred)
    print("r2 score is: {}".format(r2score))
    print('MAE:{}'.format(mean_absolute_error(y_test,pred)))
    print('MSE:{}'.format(mean_squared_error(y_test,pred)))
    print('RMSE:{}'.format(np.sqrt(mean_squared_error(y_test,pred))))
        
```

```python
from xgboost import XGBRegressor
xgb = XGBRegressor(learning_rate=0.02, 
                   gamma=100, 
                   max_depth=25,  
                   min_child_weight=1, 
                   max_delta_step=0, )
xgb.fit(X_train, y_train)
xgb_pred = xgb.predict(X_test)
r2score=r2_score(y_test,xgb_pred)
score=xgb.score(X_train,y_train)
print(score,xgb,':')
print("r2 score is: {}".format(r2score))
print('MAE:{}'.format(mean_absolute_error(y_test,xgb_pred)))
print('MSE:{}'.format(mean_squared_error(y_test,xgb_pred)))
print('RMSE:{}'.format(np.sqrt(mean_squared_error(y_test,xgb_pred))))
        
```
