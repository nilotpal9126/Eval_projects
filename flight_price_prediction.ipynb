It can be seen that the dataset contains columns like 'Duration', 'Dep_time', 'Arrival time' and 'Date of journey' that need to be converted to date time format.

```python
import datetime as dt
```

```python
df1['Dep_Time'] = pd.to_datetime(df1['Dep_Time']) #changing the dtpye for departure and arrival time 
df1['Arrival_Time'] = pd.to_datetime(df1['Arrival_Time'])
```

```python
df1['year'] = pd.DatetimeIndex(df1['Date_of_Journey']).year 
df1['month'] = pd.DatetimeIndex(df1['Date_of_Journey']).month 
df1['day'] = pd.DatetimeIndex(df1['Date_of_Journey']).day 
#as the datatype has changed now the hours and minutes can be further extracted tyo get better understanding of the data
df1['Dep_time(hrs)']= df1['Dep_Time'].dt.hour
df1['Dep_time(mins)']= df1['Dep_Time'].dt.minute
df1['Arrival_time(hrs)']= df1['Arrival_Time'].dt.hour
df1['Arrival_time(mins)']= df1['Arrival_Time'].dt.minute
```

```python
ds1['Dep_Time'] = pd.to_datetime(ds1['Dep_Time']) #changing the dtpye for departure and arrival time 
ds1['Arrival_Time'] = pd.to_datetime(ds1['Arrival_Time'])
ds1['year'] = pd.DatetimeIndex(ds1['Date_of_Journey']).year 
ds1['month'] = pd.DatetimeIndex(ds1['Date_of_Journey']).month 
ds1['day'] = pd.DatetimeIndex(ds1['Date_of_Journey']).day 
#as the datatype has changed now the hours and minutes can be further extracted tyo get better understanding of the data
ds1['Dep_time(hrs)']= ds1['Dep_Time'].dt.hour
ds1['Dep_time(mins)']= ds1['Dep_Time'].dt.minute
ds1['Arrival_time(hrs)']= ds1['Arrival_Time'].dt.hour
ds1['Arrival_time(mins)']= ds1['Arrival_Time'].dt.minute
```

```python
#Seperatiing into x and y(for train)
x=df1.drop(['Price'],axis=1)
y=df1['Price']
```

```python
from sklearn.preprocessing import MinMaxScaler
MM=MinMaxScaler()
MM_fit=MM.fit(x)
X=MM_fit.transform(x)
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression  
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingRegressor,RandomForestRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
models=[LinearRegression(),KNeighborsRegressor(n_neighbors=5),DecisionTreeRegressor(),RandomForestRegressor(n_estimators=100,max_features='auto'),GradientBoostingRegressor()]
for m in models:
    m.fit(X_train,y_train)
    y_pred=m.predict(X_test)
    score=m.score(X_train,y_train)
    pred=m.predict(X_test)
    print(score,m,':')
    print("Predictions are: {}".format(pred))
    
    r2score=r2_score(y_test,pred)
    print("r2 score is: {}".format(r2score))
    print('MAE:{}'.format(mean_absolute_error(y_test,pred)))
    print('MSE:{}'.format(mean_squared_error(y_test,pred)))
    print('RMSE:{}'.format(np.sqrt(mean_squared_error(y_test,pred))))
        
```

```python
from sklearn.linear_model import Lasso,Ridge
ls=Lasso(alpha=0.0001)
ls.fit(X_train,y_train)
ls.score(X_train,y_train)
print('ls_coef',ls.coef_)
rd=Ridge(0.001)
rd.fit(X_train,y_train)
print('rd_score',rd.score(X_train,y_train))
print('rd_coef',rd.coef_)
```

#### Hyperparameter tunning

```python
from sklearn.model_selection import GridSearchCV
params= { 
    'n_estimators': [200, 500],
    'max_features': ['auto', 'sqrt', 'log2'],'criterion':['mse','mae']
}
rf=RandomForestRegressor()
rf = GridSearchCV(rf, params, cv=5)
rf.fit(X_train,y_train)
rf.best_params_   
```

```python
params= {'max_features': ['auto', 'sqrt', 'log2'],'criterion':['mse', 'friedman_mse', 'mae', 'poisson'],'splitter':['best','random']}
dtc=DecisionTreeRegressor()
dtc=GridSearchCV(dtc, params, cv=5)
dtc.fit(X_train,y_train)
dtc.best_params_
```

```python
#checkiing the model with hypertunned  parameters
rf=RandomForestRegressor(criterion= 'mse', max_features= 'auto', n_estimators= 200)
rf.fit(X_train,y_train)
y_pred=rf.predict(X_test)
rf_score=rf.score(X_train,y_train)
rf_pred=rf.predict(X_test)
print('rf_score',rf_score,':')
print("Predictions are: {}".format(rf_pred))
    
```

For the random forest the score increses after hypertunning the model.


```python
#checkiing the model with hypertunned  parameters
dtc=DecisionTreeRegressor(criterion= 'friedman_mse', max_features= 'auto',splitter= 'best' )
dtc.fit(X_train,y_train)
y_pred=dtc.predict(X_test)
dtc_score=dtc.score(X_train,y_train)
dtc_pred=dtc.predict(X_test)
print('dtc_score',dtc_score,':')
print("Predictions are: {}".format(dtc_pred))


        
```
