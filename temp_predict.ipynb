as there are two target variables so the data has to be slpit into two parts ,i.e y1 and y2 and the training/ testing/ modelling/ tunning/ prediction/ scoring of the data will be don for both the target varibales.  

```python
x=df1.drop(['Next_Tmax'],axis=1)
x=df1.drop(['Next_Tmin'],axis=1)
y1=df1['Next_Tmax']
y2=df1['Next_Tmin']
print(x,y1,y2)
```

```python
from sklearn.preprocessing import MinMaxScaler
MM=MinMaxScaler()
MM_fit=MM.fit(x)
X=MM_fit.transform(x)
print(X)






```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y1_train,y1_test=train_test_split(X,y1,test_size=0.3, random_state=42)
X_train,X_test,y2_train,y2_test=train_test_split(X,y2,test_size=0.3, random_state=42)
```

#### Model building

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y1_train) #for y1 target variable
print(lr.score(X_train,y1_train))
lr_pred1=lr.predict(X_test)
print(lr_pred1)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y2_train) #for y2 target variable
print(lr.score(X_train,y2_train))
lr_pred2=lr.predict(X_test)
print(lr_pred2)
```

#### for y1 the accuracy is 100% lr and for y2 it is 84%

```python
from sklearn.metrics import mean_squared_error
#RM= RootMeanSquaredError
M1= mean_squared_error(lr_pred1, y1_test)
M2=mean_squared_error(lr_pred2, y2_test)
RM1=np.sqrt(M1)
RM2=np.sqrt(M2)
print(M1)
print(M2)
print(RM1)
print(RM2)
```

```python
from sklearn.linear_model import Lasso,Ridge
ls=Lasso(alpha=0.0001)
ls.fit(X_train,y1_train)
print(ls.score(X_train,y1_train))
ls.fit(X_train,y2_train)
print(ls.score(X_train,y2_train))
```

```python
ls.coef_
```

```python
rd=Ridge(0.001)
rd.fit(X_train,y1_train)
print(rd.score(X_train,y1_train))
rd.fit(X_train,y2_train)
print(rd.score(X_train,y2_train))
```

```python
rd.coef_
```

```python
from sklearn.svm import SVR
KL=['linear','poly','rbf']
for i in KL:
    sv=SVR(kernel=i)
    sv.fit(X_train,y1_train)
    sv_pred1=sv.predict(X_test)
    print(sv_pred1)
    print(sv.score(X_train,y1_train))
```

```python
from sklearn.svm import SVR
KL=['linear','poly','rbf']
for i in KL:
    sv=SVR(kernel=i)
    sv.fit(X_train,y2_train)
    sv_pred2=sv.predict(X_test)
    print(sv_pred2)
    print(sv.score(X_train,y2_train))
```

```python
M1=mean_squared_error(sv_pred1,y1_test)
M2=mean_squared_error(sv_pred2,y2_test)
RM1=np.sqrt(M1)
RM2=np.sqrt(M2)
print(M1)
print(M2)
print(RM1)
print(RM2)
```
